## DATA PREPRATION ##

### 1. Binary Conversion ###

```python
months = pd.get_dummies(df.month,drop_first=True)
weekdays = pd.get_dummies(df.weekday,drop_first=True)
weathersits = pd.get_dummies(df.weathersit,drop_first=True)
seasons = pd.get_dummies(df.season,drop_first=True)

df = pd.concat([months,weekdays,weathersits,seasons,df],axis=1)
df.head()
```

```python
df.drop(['month', 'weekday', 'weathersit', 'season'], axis = 1, inplace = True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### 2. Create Train & Test Data ###

```python
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df,train_size = 0.7, test_size = 0.3, random_state=200)

print(df_train.shape, df_test.shape)
```

### 3. Undergo Rescaling - Normalization ###

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

df_num = ['temp','atemp','humidity','windspeed','count']

df_train[df_num] = scaler.fit_transform(df_train[df_num])

df_train[df_num].head()
```

### 4. Create Heat Map ###

```python
plt.figure(figsize = (25,25))
sns.heatmap(df_train.corr(), annot=True, cmap='YlGnBu')
plt.show()
```

```python
df_train.drop(['casual', 'registered'], axis = 1, inplace = True)
df_train.shape
```

### 4. Build Model - Using RFE (Recurrsive Feature Elimination) Approach ###

```python
#Create Model
from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression

y_train = df_train.pop('count')
X_train = df_train

lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm,20)
rfe =rfe.fit(X_train, y_train)
```

```python
#Show RFE Results
list(zip(X_train.columns, rfe.support_, rfe.ranking_))
```

```python
#Identify Columns where RFE is True
col = X_train.columns[rfe.support_]
print(col)
```

```python
# Drop False columns from DF and create new X_train_rfe_df
X_train_rfe_df = X_train[col]
X_train_rfe_df.head()
```

```python
print(X_train.shape)
print(X_train_rfe_df.shape)
```

#Result: 11 columns have been successfully removed

```python
import statsmodels.api as sm

# Add constant
X_train_sm = sm.add_constant(X_train_rfe_df)

# Create a LR model
lr = sm.OLS(y_train, X_train_sm)

lr_model = lr.fit()

#print(lr_model.params)
print(lr_model.summary())
```

```python
# Remove Holiday with P value > 0.05
X_train_sm1 = X_train_sm.drop(['holiday'], axis = 1)

# Create a LR model
lr1 = sm.OLS(y_train, X_train_sm1)

lr1_model = lr1.fit()

#print(lr_model1.params)
print(lr1_model.summary())
```

```python
# Remove Sunday with P value > 0.05
X_train_sm2 = X_train_sm1.drop(['Sunday'], axis = 1)

# Create a LR model
lr2 = sm.OLS(y_train, X_train_sm2)

lr2_model = lr2.fit()

#print(lr_model1.params)
print(lr2_model.summary())
```

```python
# Remove July with P value > 0.05
X_train_sm3 = X_train_sm2.drop(['Jul'], axis = 1)

# Create a LR model
lr3 = sm.OLS(y_train, X_train_sm3)

lr3_model = lr3.fit()

#print(lr_model1.params)
print(lr3_model.summary())
```

```python
#Calculate VIF for latest Model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_sm3.columns

vif['VIF'] = [variance_inflation_factor(X_train_sm3.values, i) for i in range(X_train_sm3.shape[1])]

vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)

print(vif)
```

```python
# Remove Spring with VIF > 5
X_train_sm4 = X_train_sm3.drop(['Spring'], axis = 1)

# Create a LR model
lr4 = sm.OLS(y_train, X_train_sm4)

lr4_model = lr4.fit()

#print(lr_model1.params)
print(lr4_model.summary())
```

```python
#Calculate VIF for latest Model
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = pd.DataFrame()
vif['Features'] = X_train_sm4.columns

vif['VIF'] = [variance_inflation_factor(X_train_sm4.values, i) for i in range(X_train_sm4.shape[1])]

vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)

print(vif)
```

##The F-Statistics value of 185.5 and the p-value is ~0 suggests that overall model is significant##
